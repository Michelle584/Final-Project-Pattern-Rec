```python
def train_transfer_learning_CNN(num_of_classes,save_file_as,data_train, labels_train, img_size=214, epochs=5, learning_rate=1e-4):
    
    # Normalize data
    labels_train = np_utils.to_categorical(labels_train, num_of_classes)
    
    # specify a new input shape to replace VGG16 input layer for our data
    new_input = (img_size, img_size, 3)
    
    # load model
    model = VGG16(include_top=False, input_shape=new_input, pooling='max')
    
    #remove output layer to replace with new one to allow for classification of the 3 classes instead of the original VGG16
    # classification 
    flat1 = Flatten()(model.layers[-1].output) # remove output layer
    class1 = Dense(1024, activation='relu')(flat1) # new dense layer
    output = Dense(num_of_classes, activation='softmax')(class1) # for 3 classes
    
    # define new model
    model = Model(inputs=model.inputs, outputs=output)
    
    # print summary of model
    model.summary()

    # set Adam learning rate
    adam = Adam(lr=learning_rate)

    # We add metrics to get more results you want to see
    model.compile(optimizer=adam,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # fit model
    print('Training ------------')
    history = model.fit(data_train,labels_train, epochs=epochs)
                
    # plot training accuracy
    plt.plot(history.history['accuracy'])
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epochs')
    plt.legend(['Training'], loc='upper left')
    plt.show()

    # save trained VGG16 CNN model
    model.save(save_file_as)
```

```python
def train_scratch_CNN(num_of_classes,save_file_as,data_train,labels_train, img_size=100, epochs=15, batch_size=20, learning_rate=1e-4):
    

    
    model = Sequential()

    model.add(Convolution2D(filters = 64, kernel_size = 2, activation ='relu', input_shape =(img_size,img_size,3), name = 'Layer1'))
    model.add(Convolution2D(filters = 64, kernel_size = 2, activation ='relu' , name = 'Layer2'))
    model.add(Convolution2D(filters = 64, kernel_size = 2, activation ='relu' , name = 'Layer3'))
    model.add(Convolution2D(filters = 64, kernel_size = 2, activation ='relu' , name = 'Layer4'))
    
    model.add(MaxPooling2D(pool_size = 2, name = 'MaxPooling_Layer1'))
    model.add(Dropout(0.5, name = 'Dropout1'))

    model.add(Convolution2D(filters = 32, kernel_size = 2, activation ='relu', name = 'Layer5'))
    model.add(Convolution2D(filters = 32, kernel_size = 2, activation ='relu', name = 'Layer6'))
    model.add(Convolution2D(filters = 32, kernel_size = 2, activation ='relu', name = 'Layer7'))

    model.add(MaxPooling2D(pool_size = 2, name = 'MaxPooling_Layer2'))
    model.add(Dropout(0.5, name = 'Dropout2'))

    model.add(Flatten(name = 'Flatten'))
    model.add(Dense(128, activation='relu', name = 'DenseLayer'))

    model.add(Dense(num_of_classes, activation='softmax', name = 'OutputLayer'))

    y_train = np_utils.to_categorical(labels_train, num_classes=num_of_classes)
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    # fit model 
    print('Training ------------')
    history = model.fit(data_train, y_train, epochs=epochs, batch_size=batch_size)
    
    # plot training accuracy
    plt.plot(history.history['accuracy'])
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epochs')
    plt.legend(['Training'], loc='upper left')
    plt.show()
    
    model.save('scratch_trained_cnn.hdf5')
```

## Training on all images with CNN from scratch

```python
if os.path.isfile('scratch_trained_cnn.hdf5')==False:
    train_scratch_CNN(num_of_output_classes,'scratch_trained_cnn.hdf5',X_train_preprocess,t_train)
```

## Training on all images with transfer learning

```python
# train VGG16 CNN if there is no file
if os.path.isfile('vgg16_trained_cnn.hdf5')==False:
    train_transfer_learning_CNN(num_of_output_classes,'vgg16_trained_cnn.hdf5',X_train_preprocess,t_train, img_size=100, epochs=5, learning_rate=1e-4)
```
